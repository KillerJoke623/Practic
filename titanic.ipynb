```python
train_data = pd.read_csv("../input/titanic/train.csv")
test_data = pd.read_csv("../input/titanic/test.csv")
```

```python
women = train_data.loc[train_data.Sex == 'female']["Survived"]
rate_women = sum(women)/len(women)

print("% of women who survived:", rate_women*100)
```

```python
men = train_data.loc[train_data.Sex == 'male']["Survived"]
rate_men = sum(men)/len(men)

print("% of men who survived:", rate_men*100)
```

```python
new_df = train_data.query('Survived == 1') \
    .groupby('Sex', as_index = False) \
    .agg({'Name' : 'nunique'})
new_df2 = train_data.groupby('Sex', as_index = False) \
    .agg({'Name' : 'nunique'})
new_df = new_df.merge(new_df2, on='Sex')
new_df = new_df.rename(columns={'Name_x' : 'survived', 'Name_y' : 'total', 'Sex' : 'sex'})
```

```python
from sklearn.ensemble import RandomForestClassifier

y = train_data["Survived"]

features = ["Pclass", "Sex", "SibSp", "Parch"]
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X, y)
predictions = model.predict(X_test)

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
export_csv = output.to_csv(r'domashnee.csv', index=None, header=True)
```
